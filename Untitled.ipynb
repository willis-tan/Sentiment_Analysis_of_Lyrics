```python
import re
import os

for artist in artists:
    f = open('lyrics_' + artist + ".txt", 'rb')
    all_words = ''
    for sentence in f.readlines():
        this_sentence = sentence.decode('utf-8')
        all_words += this_sentence
    f.close()
    
    #remove identifiers like chorus, verse, etc
    pattern = '(Lyrics for this song have yet to be released\. Please check back once the song has been released\.\s)'\
              '|(\s+<\|endoftext\|>\s{5})'
    all_words = re.sub(pattern, '', all_words)
    all_words = re.sub(r'-\s{1}', '', all_words)
    #remove empty lines
    all_words = os.linesep.join([s for s in all_words.splitlines() if s])
    
    f = open('lyrics_' + artist + '-cleaned.txt', 'wb')
    f.write(all_words.encode('utf-8'))
    f.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)

from nltk.corpus import stopwords
from wordcloud import WordCloud

SW= stopwords.words('english')

df = pd.DataFrame(columns=('artist', 'lexicalrichness'))
for artist in artists:
    f = open('lyrics_' + artist + '-cleaned.txt', 'rb')
    all_words = ''
    raw_text = ""
    for sentence in f.readlines():
        this_sentence = sentence.decode('utf-8')
        raw_text += this_sentence.lower()
    
    words = raw_text.split()
    filtered_words = [word for word in words if word not in SW
                      and len(word) > 1] # remove the stopwords
    
    # generate wordclouds
    word_cloud = WordCloud(width=1000, height=500).generate(' '.join(filtered_words))
    word_cloud.to_file('wordclouds_' + artist + ".png")
    image = word_cloud.to_image()
    
    df.loc[i] = (artist, 100*len(set(filtered_words)) / len(filtered_words))
    i+=1
    
df.plot.bar(x='artist', y='lexicalrichness', title='Lexical richness of each Artist');
plt.show()
```
