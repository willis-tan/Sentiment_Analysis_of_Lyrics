# Cleaning the Data

```python
import re

artists = ['Deafheaven', 'Beastie Boys', 'Grouplove', 'Rage Against the Machine', 'The Killers',
           'Linkin Park', 'Morbid Angel', 'Smashing Pumpkins', 'Gorillaz', 'Foo Fighters']
for artist in artists:
    path = os.path.join(root, 'lyrics_' + artist + '.txt')
    f = open(path, 'rb')
    all_words = ''
    for sentence in f.readlines():
        this_sentence = sentence.decode('utf-8')
        all_words += this_sentence
    f.close()
    
    # remove endoftext delimiters, unreleased lyrics message, and whitespace
    pattern = '(Lyrics for this song have yet to be released\. Please check back once the song has been released\.\s)'\
              '|(\s+<\|endoftext\|>\s{5})'
    all_words = re.sub(pattern, '', all_words)
    all_words = re.sub(r'-\s{1}', '', all_words)
    # remove empty lines
    all_words = os.linesep.join([s for s in all_words.splitlines() if s])
    # write cleaned lyrics to new file
    path = os.path.join(root, 'lyrics_' + artist + '-cleaned.txt')
    f = open(path, 'wb')
    f.write(all_words.lower().encode('utf-8'))
    f.close()
```

## Lexical Richness of Each Artist

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()

# function to convert nltk tag to wordnet tag
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

def lemmatize_sentence(sentence):
    #tokenize the sentence and find the POS tag for each token
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))  
    #tuple of (token, wordnet_tag)
    wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
        if tag is None:
            #if there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:        
            #else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)

from stopwords import stopwords as SW
from wordcloud import WordCloud

i = 0
df = pd.DataFrame(columns=('artist', 'lexicalrichness'))
for artist in artists:
    path = os.path.join(root, 'lyrics_' + artist + '-cleaned.txt')
    f = open(path, 'rb')
    raw_text = ""
    for sentence in f.readlines():
        this_sentence = sentence.decode('utf-8')
        raw_text += this_sentence
    
    words = raw_text.split()
    filtered_words = [word for word in words if word not in SW
                      and len(word) > 1] # remove the stopwords from raw text
    
    # generate wordclouds
    path = os.path.join('clouds', 'wordclouds_' + artist + ".png")
    word_cloud = WordCloud(width=1000, height=500).generate(lemmatize_sentence(' '.join(filtered_words)))
    word_cloud.to_file(path)
    
    df.loc[i] = (artist, 100*len(set(filtered_words)) / len(filtered_words))
    i+=1

df.sort_values('lexicalrichness').plot.bar(x='artist', y='lexicalrichness', title='Lexical richness of each Artist');
plt.show()
```

From the bar plot, we can clearly see that Deafheaven has the highest amount of lexical richness out of all the artists. Their lyrics tend to be very poetic and abstract, so that's one factor. The biggest contributor would be the use of spoken word lyrics.

# Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

df = pd.DataFrame(columns=('artist', 'pos', 'neu', 'neg'))
sia = SentimentIntensityAnalyzer()
i=0
for artist in artists:
    num_positive = 0
    num_negative = 0
    num_neutral = 0

    path = os.path.join(root, 'lyrics_' + artist + '-cleaned.txt')
    f = open(path, 'rb')
    for sentence in f.readlines():
        this_sentence = sentence.decode('utf-8')
        score = sia.polarity_scores(this_sentence)
        score = score['compound']
        # classify whether a sentence is positive, neutral, or negative
        if score >= 0.05:
            num_positive += 1
        elif score > -0.5 and score < 0.5:
            num_neutral += 1
        else:
            num_negative += 1
    
    # get the % of positive, neutral, or negative sentences per artist
    num_total = num_negative + num_neutral + num_positive
    percent_negative = (num_negative / float(num_total)) * 100
    percent_neutral = (num_neutral / float(num_total)) * 100
    percent_positive = (num_positive / float(num_total)) * 100
    print(artist, percent_negative, percent_neutral, percent_positive)
    df.loc[i] = (artist, percent_positive, percent_neutral, percent_negative)
    i+=1
    f.close()

df.plot.bar(x = 'artist', stacked=True)
plt.show()
```
